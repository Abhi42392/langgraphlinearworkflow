```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv
from langchain_anthropic import ChatAnthropic
```

```python
class BlogState(TypedDict):
    title:str
    outline:str
    blog:str
```

```python
load_dotenv()
```

```python
model=ChatAnthropic(model="claude-sonnet-4-6")
```

```python
def outline_generator(state):
    title=state['title']
    title=f'Generate a simple blog outline for this {title}'
    response=model.invoke(title).content
    state['outline']=response
    return state
```

```python
def blog_generator(state):
    title=state['title']
    outline=state['outline']
    request=f'generate short blog for the title {title} and outline of the blog {outline}'
    response=model.invoke(request).content
    state['blog']=response
    return state
```

```python
graph=StateGraph(BlogState)
graph.add_node('outline_generator',outline_generator)
graph.add_node('blog_generator',blog_generator)
graph.add_edge(START,'outline_generator')
graph.add_edge('outline_generator','blog_generator')
graph.add_edge('blog_generator',END)
workflow=graph.compile()
```

```python
initial_state={'title':'Agentic AI'}
final_state=workflow.invoke(initial_state)
```

```python
print(final_state['blog'])
```
